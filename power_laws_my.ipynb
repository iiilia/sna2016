```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.random as rnd
%matplotlib inline
```

```python
r = rnd.random(10000)
alpha = 3.5
xmin = 1.0;

x = (1 - r)**(-1.0/(alpha - 1)) * xmin
```

```python
yh, binEdges = np.histogram(x,bins=1000)
bincenters = 0.5*(binEdges[1:]+binEdges[:-1])
plt.plot(bincenters, yh, lw=2)
plt.ylabel('count')
plt.xlabel('x')
```

```python
# cdf
Fx = 1- np.power((x/xmin),1-alpha)
plt.plot(x,Fx,'.')
```

```python
plt.loglog(x,1-Fx,'.')
```
